```python
score_idorder_df = pd.merge(
    left=order_items_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
score_idorder_df.head()
```

```python
productid_score =  pd.merge(
    left=product_df,
    right=score_idorder_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
productid_score.head()
```

```python
eng_productid_score =  pd.merge(
    left=product_category_name_df,
    right=productid_score,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
eng_productid_score.head()
```

```python
score_product = eng_productid_score.groupby(['product_category_name_english'])['review_score'].mean()
resetindex = score_product.sort_values(ascending=False).reset_index().head(5)
resetindex
```
